```python
class length:
    def inp(self):
        print("Enter length of the sides of triangle: ")
        self.a=int(input())
        self.b=int(input())
        self.c=int(input())
        
class area(length):
    def cal_area(self):
        self.inp()
        self.s=(self.a+self.b+self.c)/2
        area = (self.s*(self.s-self.a)*(self.s-self.b)*(self.s-self.c))**0.5
        return area
    
tri = area()
a=tri.cal_area()
print("Area of triangle: ",a)
```

```python
def filter_long_words(l,n):
    l1=[]
    for i in l:
        if len(i)>n:
            l1.append(i)
    return l1

no=int(input("Enter number of words: "))
l=[]
for i in range(no):
    l.append(input())
n=int(input("Enter n: "))
print("List of words longer than n is: ",filter_long_words(l,n))
```

```python
def length(l):
    return [len(i) for i in l]

no=int(input("Enter number of words: "))
l=[]
for i in range(no):
    l.append(input())
print("Length of words: ",length(l))
```

```python
def vowel(s):
    if s.lower() in 'aeiou':
        return True
    else:
        return False
        
s=input("Enter a character: ")
vowel(s)
```
